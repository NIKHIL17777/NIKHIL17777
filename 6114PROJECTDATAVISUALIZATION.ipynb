```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create a DataFrame from a CSV file
df = pd.read_csv('/content/auction_2023 (1).csv')  # Replace 'your_data.csv' with the actual CSV file path

# Step 2: Clean the 'Price' column and extract numeric values
# Assuming 'Price' column contains strings with currency symbols and other text
df['Price'] = df['Price'].str.replace('[^\d.]', '', regex=True).astype(float)

# Step 3: Descriptive statistics
mean = df['Price'].mean()
median = df['Price'].median()
mode = df['Price'].mode().values[0]
std_dev = df['Price'].std()

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev}")

# Step 4: Generate a count plot for 'Price' scores
plt.figure(figsize=(22, 6))
sns.countplot(x='Price', data=df, color='red')
plt.title("The Price")
plt.xlabel("Price Scores")
plt.ylabel("Count")
plt.show()

# Step 5: Generate a histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Price'], bins=4, color='red', edgecolor='black')
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

team_specific_palette = {
    'UP Warriorz': '#50276E',
    'Mumbai Indians': '#045093',
    'Delhi Capitals': '#17449B',
    'Gujarat Giants': '#FDA736',
    'Royal Challengers Bangalore': '#DA1818'
}

fig = plt.figure(figsize=(20, 6))

players_per_team = df.groupby('Team')['Player'].count().reset_index().sort_values(by='Player', ascending=True)

ax = sns.barplot(y='Team', x='Player', data=players_per_team, palette=list(team_specific_palette.values()))
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=12, color='black', padding=3, weight='bold')
plt.title('Number of Players in Each Team', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('')
plt.xlabel('Number of Players', fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Your existing code for data preparation
money_spent_by_each_team = df.groupby('Team')['Price (US$)'].sum().reset_index().sort_values(by='Price (US$)', ascending=True)
money_spent_by_each_team['Price (US$)'] = money_spent_by_each_team['Price (US$)'] / 1000000
money_spent_by_each_team['Price (US$) Label'] = money_spent_by_each_team['Price (US$)'].apply(lambda x: f'${x:.2f} M')

# Create a radar chart
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, polar=True)

# Number of teams
num_teams = len(money_spent_by_each_team)

# Angles for each axis (equally spaced)
angles = np.linspace(0, 2 * np.pi, num_teams, endpoint=False).tolist()
angles += angles[:1]  # To close the circle

# Money spent data
money_spent_data = money_spent_by_each_team['Price (US$)'].tolist()
money_spent_data += money_spent_data[:1]  # To close the circle

# Convert angles to degrees
angles_degrees = [angle * 180/np.pi for angle in angles]

# Plot the radar chart
ax.fill(angles, money_spent_data, 'b', alpha=0.1)
ax.set_thetagrids(angles_degrees[:-1], money_spent_by_each_team['Team'].tolist(), fontsize=12)
ax.set_title('Total Money spent by each team (US$)', fontsize=16)

plt.show()
```
